```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

label = LabelEncoder()

# 法一：
te['platform'] = label.fit_transform(tr['platform'])  
te['media_name'] = label.fit_transform(tr['media_name']) 
te
```

```python
equa = pd.read_csv("模型拟合优度.csv")
```

```python
equa
```

```python
df3 = pd.merge(te,equa,on=['platform','media_name'],how='left')
df3 = df3.drop(columns = 'fitness')
```

```python
df3
```

```python
curve_reach = []
for row in df3.index:
    equation0 = df3.loc[row]['equation']
    X0 = df3.loc[row]['grp']
    Y_predict = list(map(lambda x:eval(equation0),[X0]))
    for i in Y_predict:
        curve_reach.append(i)


curve_reach = pd.DataFrame(curve_reach)   
        
curve_reach[curve_reach < 0] = 0
```

```python
prediction = pd.concat([te,curve_reach],axis=1,ignore_index=False)
prediction.rename(columns = { 0 : 'curve_reach'},inplace = True)
prediction[u'curve_reach'] = prediction[u'curve_reach'].apply(lambda x: format(x, '.2%'))  
prediction
```

```python
prediction.to_csv('Test_Prediction0.csv')
```
